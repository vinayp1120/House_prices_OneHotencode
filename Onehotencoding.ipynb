```python
import pandas as pd
df=pd.read_csv("/content/homeprices.csv")
print(df)
```

Now as we see that town is categorical column we converted them into numerical values with the help of dummy variables.we converted them to numerical and made dummy variables and stored it next step is concataneate this with the original data

```python
dummies=pd.get_dummies(df.town,dtype=int)
dummies
```

Concatenate them using pd.concat() list of dataframes to concatenate and how u like to add them like row wise or colum wise

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

Multicollinearity:
Multicollinearity means two or more variables in your model are too similar (highly correlated). When this happens, the model gets confused because it can't tell which variable is more important in explaining the result. It makes the model less accurate and harder to interpret.

Example: If you have two variables: Height and Weight, and they are very similar, your model will struggle to understand if changes in height or weight are affecting the outcome because they are closely related. This is multicollinearity.

The dummy variable trap happens when you create dummy variables for all categories in a categorical variable, leading to redundant information. This makes one variable perfectly predictable from others, causing issues in your model. To avoid it, leave out one dummy variable as a reference (baseline) category.

```python
final_dataset=merged.drop(["town","west windsor"],axis="columns")
final_dataset
```

Now lets make a linear regression model using sklearn module

```python
from inspect import modulesbyfile
from sklearn.linear_model import LinearRegression
#now for the class Linear regression lets make an instance i.e an object to use the linear regression
model = LinearRegression()
```

Now its time to create the input data for the Linear model.Now our aim is to predict the prices

```python
x= final_dataset.drop(["price"],axis='columns')
x
```

```python
y=final_dataset['price']
y
```

Now lets try to train the model ok 🎉🎉🎉

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

Now lets Find out the accuracy of the model using Score method.if takes input and actula output and the predicted value by the model is compared with the actual output and with the help of a formula it calculates its accuracy:😊😊

```python
model.score(x,y)
```

Now lets use the method One Hot encoder for the categorical values before we used the dummy variables

```python
from sklearn.preprocessing import LabelEncoder
# create an object for the Lable Encoder and store the object
le=LabelEncoder()
```

```python
#df is the dataframe we stored our table or dataset
#we changed the categorical values and assgined back to the dataframe and see the output table
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
#this creates a 2d array with values of [town,area] as a list
X=dfle[['town','area']].values
X
```

```python
y=dfle['price']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assume your data is stored in 'X'
# Apply OneHotEncoder to the 0th column (categorical column)
column_transformer = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [0])  # Apply OneHotEncoder to the 0th column
    ],
    remainder='passthrough'  # Keep other columns unchanged
)
```

```python
X_transformed = column_transformer.fit_transform(X)
X_transformed
```

```python
#drop one column to avoid the dummy trap
X_transformed=X_transformed[:,1:]
X_transformed
```

```python
model.fit(X_transformed,y)
```

```python
model.score(X_transformed,y)
```
